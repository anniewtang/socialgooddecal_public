```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_data(k, mu = 3):
    noise = np.random.normal(0,1,k)
    true_data = mu*np.ones(k)
    noisy_data = true_data + noise
    return(noisy_data)
```

```python
means = []
for i in range(1,500):
    means.append(np.mean(gen_data(i)))
plt.plot(means)
```
